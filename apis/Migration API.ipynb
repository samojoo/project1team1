```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
url = "https://open.alberta.ca/dataset/eece1571-4b82-40be-8a11-4627db1a9561/resource/2acfe7ec-6c54-48ea-bcb8-6aa72fd64a7a/download/rows"
response = requests.get(url).json()
pprint (response)
```

```python
#pprint (response['data'][6])
net_migration =[]
time_period = []
for index in range(0,831,3):
    net_migration.append(response['data'][index][12])

for index in range(0,831,3):
    time_period.append((response['data'][index][8].split("-01",1)[0]))
```

```python
#Create Dataframe
migration_df = pd.DataFrame({"Date": time_period,"National Net Migration": net_migration})

migration_df["National Net Migration"] = pd.to_numeric(migration_df["National Net Migration"])

```

```python
#Create line graph
plt.figure(figsize=(30,10))
plt.plot(migration_df["Date"],migration_df["National Net Migration"])
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()
```

```python
# Push the remade DataFrame to a new CSV file
migration_df.to_csv('..cleaning/nationalmigration_clean.csv',
                  index=False, header=True)
```
